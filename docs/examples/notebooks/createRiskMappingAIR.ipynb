```python
import csv
import json

# Third Party
from linkml_runtime.dumpers import YAMLDumper


from risk_atlas_nexus.ai_risk_ontology.datamodel.ai_risk_ontology import Risk, Container
from risk_atlas_nexus import RiskAtlasNexus
DATA_DIR = "/Users/ingevejs/Documents/workspace/ingelise/risk-atlas-nexus/src/risk_atlas_nexus/data/knowledge_graph/"

ran = RiskAtlasNexus() 

def create_container_object() -> Container:

    # Add risk-related actions to risks
    risk_data = get_risks()
    # Create container
    container = Container(risks=risk_data)
    return container

not_found = []
def parse_risks(related_risks: str):
    #print(related_risks)
    #print(type(related_risks))
    risk_ids = []
    
    if type(related_risks) is str:
        rr= json.loads(related_risks)
    else:
        rr = related_risks 
    #print(rr)
    for item in rr:
        #print(item)
        risk = ran.get_risk_by_name(item)
        if risk:
            risk_ids.append(risk.id)
        else: 
            not_found.append(item)
    return risk_ids

def get_risks():
    with open("../../resources/sample.csv") as csvfile:
        importer = csv.DictReader(csvfile)
        the_sample = [row for row in importer]

    risk_objects = []

    for row in the_sample:
        #print(row["l4-name"])
        risk_objects.append(Risk(
            **{
                "id": "ai-risk-taxonomy-" + row["l4-name"].lower().replace(" ", "-"),
                "relatedMatch": parse_risks(row["related-risk-atlas"])
            }))

    print("Not found", not_found)
    return(risk_objects)

container = create_container_object()
with open(DATA_DIR + "air_to_ibm_risk_atlas.yaml", "+tw", encoding="utf-8") as output_file:
    print(YAMLDumper().dumps(container), file=output_file)
    output_file.close()
```
